# Visão Geral dos Dados - Clinic Finances

```python
import sys
sys.path.insert(0, "..")

import pandas as pd
import matplotlib.pyplot as plt

from core import db

engine = db.engine
plt.style.use("ggplot")
plt.rcParams["figure.figsize"] = (10, 5)
```

## Carregar dados

```python
patients_df = pd.read_sql("SELECT * FROM patients", engine)
claims_df = pd.read_sql("SELECT * FROM insurance_claims", engine)
```

## Resumo de Pacientes

```python
print(f"Total de pacientes: {len(patients_df)}")
patients_df.head(10)
```

## Resumo de Guias

```python
print(f"Total de guias: {len(claims_df)}")
claims_df.head(10)
```

```python
claims_df.info()
```

## Distribuição por Status da Guia

```python
status_counts = claims_df["st_claim"].value_counts()
print(status_counts)

status_counts.plot(kind="bar", title="Distribuição por Status")
plt.xlabel("Status")
plt.ylabel("Quantidade")
plt.tight_layout()
plt.show()
```

## Distribuição por Tipo de Guia

```python
type_counts = claims_df["tp_claim"].value_counts()
print(type_counts)

type_counts.plot(kind="bar", title="Distribuição por Tipo de Guia")
plt.xlabel("Tipo")
plt.ylabel("Quantidade")
plt.tight_layout()
plt.show()
```

## Top 10 Procedimentos Mais Frequentes

```python
top_procedures = claims_df["nm_procedure"].value_counts().head(10)
print(top_procedures)

top_procedures.plot(kind="barh", title="Top 10 Procedimentos")
plt.xlabel("Quantidade")
plt.ylabel("Procedimento")
plt.tight_layout()
plt.show()
```

## Retornos vs Primeira Consulta

```python
return_counts = claims_df["fg_return"].value_counts()
return_counts.index = return_counts.index.map({True: "Retorno", False: "Primeira Consulta"})
print(return_counts)

return_counts.plot(kind="pie", autopct="%1.1f%%", title="Retornos vs Primeira Consulta")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

## Distribuição de Guias por Paciente

```python
claims_per_patient = claims_df.groupby("cd_patient").size()
print(claims_per_patient.describe())

claims_per_patient.plot(kind="hist", bins=20, title="Distribuição de Guias por Paciente", edgecolor="black")
plt.xlabel("Quantidade de Guias")
plt.ylabel("Quantidade de Pacientes")
plt.tight_layout()
plt.show()
```
